### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python
display(df.head(3))

# Nominal, Ordinal and Binary are Qualitative attributes.
# Numeric, Discrete and Continuous are Quantative attributes.

# Nominal attributes : it is related to name like Colors names, Human names etc.
# Ordinal attributes : its contains values that have a meaningful sequence or ranking(order) between them,
#                      but the magnitude between values is not actually known. 
# Binary attributes : it has only 2 data or states like yes/no, true/false, 0/1, male/female etc.
# Numeric attributes : its represented numeric value.

#                  Numerical attributes are of 2 types, interval, and ratio/range.
#                  Interval : Measured on a scale of equal-sized units like calendar dates (5-15), weight(70-85)
#                  Ratio/Range : it is a numeric attribute with a fix zero-point. like weight, height, marks, age etc.


# Here in this data set Name, Ticket, Embarked and Cabin are Nominal attribute,
# Pclass is Ordinal attribute, Survived & Sex are Binary attributes,
# Pclass, SibSp, Parch & Fare are Numeric attributes

# Unknown Column is also nominal attributes because its contains value not represent any class or any other value.
# Like 1,2,3 => 1 is Best, 2 is Average and 3 is dull
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.

```python
from pandas.api.types import is_numeric_dtype

for column in df.columns:
    if is_numeric_dtype(df[column]):
        print(column, ":")
        print("\t Mean = %.2f" % df[column].mean())
        print("\t Standard deviation = %.2f" % df[column].std())
        print("\t Minimum = %.2f" % df[column].min())
        if column != 'PassengerId' : print("\t Mode = %.2f" % df[column].mode())
        print("\t Range = %.2f" % (df[column].max() - df[column].min()))
        print("\t Maximum = %.2f" % df[column].max())
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.
