```python
df2 = df.fillna({'Age': int(df.Age.mean())})
display(df2)
```

### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
# With Min-Max

#            (age - minA)    
#  Formula = -------------  x (newMax - newMin) + newMin
#            (maxA - minA)

minA = df2.Age.min()
maxA = df2.Age.max()
newMax = 1
newMin = 0

df2["NewAge"] = ((df2.Age - minA) / (maxA - minA) * (newMax - newMin) + newMin)

display(df2)
```

```python
# With Decimal Scaling

#                     age
# Formula = ------------------------         
#            10 ^ digitCountOfMaxAge

maximumAge = df2.Age.max()
digitCount = 0

while maximumAge > 0:
    digitCount += 1
    maximumAge //= 10
    
df2["NewAge"] = (df2.Age)/(10 ** digitCount)

display(df2)
```

```python
# With Z-Score

#                age - mean()
# Formula =  ---------------------
#             StandardDeviation()

meanOfAge = df2.Age.mean()
sdOfAge = df2.Age.std()

df2["NewAge"] = (df2.Age - meanOfAge) / sdOfAge

display(df2)
```
